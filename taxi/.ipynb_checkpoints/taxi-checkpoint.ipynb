### Тренд и сезонность

```python
decomposed = seasonal_decompose(data)
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend')

plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality')

plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals')
plt.tight_layout()
```

Есть три вида сезонности:
1. дневная сезонность - данные собраны по дням, т.е. будни и выходные имеют разные продажи
2. внутрисуточная сезонность: когда продажи утром отличаются от продаж вечером
3. месячная сезонность: когда продажи лета и осени - разные.  

Для месячной сезонности у нас недостаточно данных. Поэтому посмотрим сезонность по дням. Возьмем, например, июнь.

```python
plt.figure(figsize=(12, 8))
decomposed_plot = seasonal_decompose(data_plot['num_orders']['2018-06-01':'2018-06-30'])
decomposed_plot.seasonal.plot(ax=plt.gca())
plt.title('Seasonality month')
```

## Обучение моделей

Добавим признаков в данные. Напишем функцию для этого.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour    
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

```python
make_features(data, 1, 24)
```

```python
data.head()
```

Поделим данные на обучающую и тестовую выборку, а так же разделим целевой показатель и признаки.

```python
train, test = train_test_split(data, shuffle=False, test_size=0.1)
train = train.dropna()

target_train = train['num_orders']
features_train = train.drop(['num_orders'],axis=1)
target_test = test['num_orders']
features_test = test.drop(['num_orders'],axis=1)
```

```python
features_train.shape, features_test.shape
```

Метрика для проверки качества.

```python
def rmse_metric(target, predictions):
    return (mean_squared_error(target, predictions) ** 0.5).round(2)
```

### Дерево решений

```python
dtr = DecisionTreeRegressor(random_state=12345)
parametrs = {'max_depth': range (1, 13, 2), 
             'min_samples_leaf': range (1, 8),
             'min_samples_split': range (2, 11, 2)}

tscv = TimeSeriesSplit(n_splits=8)
model_dtr = GridSearchCV(dtr, param_grid=parametrs,scoring='neg_root_mean_squared_error', n_jobs=-1, cv=tscv, verbose=1)
model_dtr.fit(features_train, target_train)

model_dtr.best_params_
```

```python
rmse_d = round(abs(model_dtr.best_score_),2)
rmse_d
```

Построим график предсказания и реальных значений.

```python
data_pred = features_test.copy()
data_pred.head()
```

```python
def draw_plt(model, day=False):
    if day == False:
        plt.figure(figsize=(13, 8))
        plt.plot(data_pred.index, model.predict(features_test),\
                 label='pred', color='darkblue')
        plt.plot(data_pred.index, target_test, label='target',alpha=0.5)
        plt.xticks(rotation = 25)
        plt.legend()
        plt.show()
    else:
        plt.figure(figsize=(13, 8))
        plt.plot(data_pred['2018-08-29':'2018-08-31'].index, \
                 model.predict(features_test['2018-08-29':'2018-08-31']), \
                 label='pred', color='darkblue')
        plt.plot(data_pred['2018-08-29':'2018-08-31'].index, \
                 target_test['2018-08-29':'2018-08-31'], \
                 label='target',alpha=0.5)
        plt.xticks(rotation = 25)
        plt.legend()
        plt.show()
```

```python
draw_plt(model_dtr.best_estimator_, True)
```

Модель предсказывает примерно в одном диапозоне, но реальные значения имеют выбросы в некоторые дни с большим количеством заказов. При этом минимум заказов почти везде предсказан верно.

### Линейная регрессия

```python
lr = linear_model.RidgeCV(scoring='neg_root_mean_squared_error')
lr.fit(features_train, target_train)

rmse_lr = round(abs(cross_val_score(lr,features_train, target_train, 
                         cv = TimeSeriesSplit(n_splits=8), 
                         n_jobs = -1, 
                         scoring = 'neg_root_mean_squared_error').mean()),2)

print(f'RMSE линейной регрессии: {rmse_lr}')                         
```

```python
draw_plt(lr)
```

Предсказания где-то по середине реальных значений.

### LightGBMRegressor

```python
lgbm = lgbm.sklearn.LGBMRegressor(random_state=12345)

param_set = {'n_estimators': range(50, 351, 50),
              'max_depth': range(1, 10, 2)   
}

tscv = TimeSeriesSplit(n_splits=8)
model_lgbm = GridSearchCV(lgbm, param_grid=param_set, scoring='neg_root_mean_squared_error', \
                          n_jobs=-1, cv=tscv, verbose=1)
model_lgbm.fit(features_train, target_train)

model_lgbm.best_params_
```

```python
rmse_lgbm = round(abs(model_lgbm.best_score_),2)
rmse_lgbm
```

```python
draw_plt(model_lgbm.best_estimator_)
```

Предсказывает количество заказов меньше чем реальные значения.

### Выбор модели

```python
d = {'RMSE':[rmse_d, rmse_lr, rmse_lgbm]}
answer = pd.DataFrame(data =d, index=['DecisionTreeRegressor','LinearRegression', 'LGBMRegressor'])
answer
```

Лучшее качество показала модель `LGBMRegressor`. Используем ее для проверки на тестовой выборки.

## Проверка на тестовой выборке

```python
rmse_metric(target_test, model_lgbm.predict(features_test))
```
